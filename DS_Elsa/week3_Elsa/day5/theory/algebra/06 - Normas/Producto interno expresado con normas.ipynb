```python
%matplotlib notebook
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# En la clase anterior vimos que la Norma L2^2 podia expresarse como el producto interno de x_t*x y que esto tiene
# beneficios computacionales.
```

```python
# Ahora veremos que el producto interno de dos vectores puede ser rescrito en funcion de la norma 2
# v1_t * v2 = norm(v1)*norm(v2)*cos(angulo) donde el angulo es el que hay entre los vectores v1 y v2 

v1 = [0,0,0,3]
v2 = [0,0,3,3]
```

```python
plt.xlim(-2, 6)
plt.ylim(-2, 6)

plt.quiver([v1[0], v2[0]],
           [v1[1], v2[1]],
           [v1[2], v2[2]],
           [v1[3], v2[3]],
           angles='xy', scale_units='xy', scale=1)

plt.show()
```

```python
# Aqui vemos que el angulo que forman el vector v1 y v2 es de 45 grados
# Calculemos v1_t * v2
v1 = np.array([0,3])
v2 = np.array([3,3])
print(v1.T.dot(v2))
```

```python
# calculemos norma(v1)*norma(v2)*cos(angulo entre v1 y v2)
v1_norma = np.linalg.norm(v1)
v2_norma = np.linalg.norm(v2)
print(v1_norma * v2_norma * np.cos(np.deg2rad(45)))
```
