```python
%matplotlib notebook
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
np.set_printoptions(suppress=True) # pedimos a numpy que no muestre los numeros pequenos de punto flotante
```

```python
# A veces necesitamos medir el tamaño de un vector, en Aprendizaje Automatico lo hacemos todo el tiempo y para
# ellos utilizamos una funcion llamada norma. Esta funcion, la norma recibe como entrada un vector y por medio
# de hacer unas cuentas le asocia a ese vector un numero no negativo, eso quiere decir que puede ser 0.

# La norma mas comun que conoces es la que mide la distancia de un punto al origen, hay muchas otras.
```

```python
# Las normas tienen las siguientes propiedades:

# 1 - Son no negativas (si las pensamos como una distancia es facil ver que no puede ser negativa)
# 2 - Es 0 si y solo si el vector es el vector 0
# 3 - Verifican la desigualdad triangular (pensando en distancia y 3 puntos tambien podemos verlo intuitivamente)
# 4 - La norma de un vector * escalar = abs(escalar) * norma(vector)
```

```python
v1 = np.array([2,7])
v2 = np.array([3,5])
print(v1)
print(v2)
```

```python
v1v2 = v1+v2
print(v1v2)
```

```python
# Calculamos la norma usando la funcion de numpy norm dentro de algebra lineal
np.linalg.norm(v1v2)
```

```python
# La desigualdad triangular dice que norm(v1+v2) <= norm(v1) + norm(v2)
norma_v1 = np.linalg.norm(v1)
norma_v2 = np.linalg.norm(v2)
norma_v1v2 = np.linalg.norm(v1v2)
print(norma_v1v2 <= (norma_v1+norma_v2))
```

```python
print(norma_v1+norma_v2)
```
