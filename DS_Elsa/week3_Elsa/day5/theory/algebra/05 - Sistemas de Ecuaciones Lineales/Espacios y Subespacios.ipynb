```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
```

```python
# En la clase anterior vimos que es una combinacion lineal de vectores y graficamos algunos puntos que podiamos
# generar como combinacion lineal de esos vectores

# En Algebra Lineal nosotros siempre estamos trabajando dentro de un espacio vectorial, es donde nuestros elementos
# viven, por ejemplo R2 o R3 que no son mas que los numeros reales en 2 o 3 dimensiones respectivamente.
```

```python
# Probemos con distintos vectores y veamos como podemos generalizar lo que graficamos
# Definamos nuevamente a v1 y v2
v1 = [1, 1]
v2 = [-1, -1]

# Grafiquemos una muestra del espacio generado  por los vectores v1 y v2
for a in range(-10, 10):
    for b in range(-10, 10):
        plt.scatter(v1[0] * a + v2[0] * b, v1[1] * a + v2[1] * b, marker='.', color='orange')

# Defining x and y sizes
plt.xlim(-25, 25)
plt.ylim(-25, 25)

# Draw axes
plt.axvline(x=0, color='#A9A9A9')
plt.axhline(y=0, color='#A9A9A9')

plt.show()
```

```python
# Recien vimos que con este eleccion de vectores estamos generando solamente una recta, si nos abstraemos de que
# estamos en R2, la recta es R y es un espacio en si mismo, pero dentro de R2, la recta es un subespacio de R2
# generado por los vectores v1 y v2
```

```python
# Probemos con distintos vectores y veamos como podemos generalizar lo que graficamos
# Definamos nuevamente a v1 y v2
v1 = [1, 0]
v2 = [2, -3]

# Grafiquemos una muestra del espacio generado  por los vectores v1 y v2
for a in range(-10, 10):
    for b in range(-10, 10):
        plt.scatter(v1[0] * a + v2[0] * b, v1[1] * a + v2[1] * b, marker='.', color='orange')

# Defining x and y sizes
plt.xlim(-50, 50)
plt.ylim(-50, 50)

# Draw axes
plt.axvline(x=0, color='#A9A9A9')
plt.axhline(y=0, color='#A9A9A9')

plt.show()
```

```python
# En este caso el espacio que podemos generar es R2 en su totalidad.

# La combinacion lineal de vectores siempre devuelve vectores que viven en el mismo espacio. En los dos ejemplos es
# lo que vimos, en el primero nuestros vectores estan sobre la misma recta, entonces no importa como los combinemos
# solamente saben responder elementos dentro de la misma recta.

# Mientras que en el segundo ejemplo los 2 vectores no estan en la misma recta, y por lo tanto son capaces de
# generar un espacio mas grande, en este caso R2, pero si usaramos esos mismos vectores en R3 estarian generando un
# hiperplano, basicamente un hiperplano siempre es un espacio de 1 dimension menos que la del espacio donde estamos
# trabajando
```

```python
# Esta importacion es para definir la proyeccion 3d, no lo utilizo para otras cosas
from mpl_toolkits.mplot3d import Axes3D

# Probemos con distintos vectores y veamos como podemos generalizar lo que graficamos
# Definamos nuevamente a v1 y v2
v1 = [1, 0, 0]
v2 = [2, -3, 0]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Grafiquemos una muestra del espacio generado  por los vectores v1 y v2
for a in range(-10, 10):
    for b in range(-10, 10):
        ax.scatter(v1[0] * a + v2[0] * b, v1[1] * a + v2[1] * b, v1[2] * a + v2[2] * b, marker='.', color='orange')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
# Podemos decir que v1 y v2 son un sistema generador del subespacio que vimos en cada caso, pero no debemos 
# confundir a los mismos con una base, ya que podrian serlo o no. Lo que si sabemos es que dentro de cada
# sistema generador podemos encontrar una base que generara lo mismo que ese sistema utilizando la menor
# cantidad posible de vectores.
```
