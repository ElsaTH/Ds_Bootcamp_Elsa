```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ahora que vimos como las combinaciones lineales de vectores nos permiten entender el espacio o subespacio
# generado por estos vectores, podemos retomar donde nos habiamos quedado, cuando un sistema de ecuaciones lineales
# tiene solucion?

# Saber si un sistema A*x = b tiene solucion se transforma ante nuestros ojos en, existe un vector x tal que la
# combinacion lineal de los vectores que componen A me permiten escribir a b. Nuestra matriz A una vez mas 
# se transformo, ahora es un sistema generador que me devuelve un subespacio y tal vez el espacio completo.
```

```python
# Retomemos el ejemplo que genera una recta en R2
# Definamos nuevamente a v1 y v2
v1 = [1, 1]
v2 = [-1, -1]

# Grafiquemos una muestra del espacio generado  por los vectores v1 y v2
for a in range(-10, 10):
    for b in range(-10, 10):
        plt.scatter(v1[0] * a + v2[0] * b, v1[1] * a + v2[1] * b, marker='.', color='orange')

# Defining x and y sizes
plt.xlim(-25, 25)
plt.ylim(-25, 25)

# Draw axes
plt.axvline(x=0, color='#A9A9A9')
plt.axhline(y=0, color='#A9A9A9')

plt.show()
```

```python
# si b = [-10,10] tiene solucion el sistema? Graficamente vemos que no, ya que el punto no pertenece al subespacio
# generado por v1 y v2. Que esta pasando? porque tengo 2 incognitas, 2 ecuaciones y aun asi el sistema no tiene 
# solucion?

# Es posible que v1 sea una combinacion lineal de v2? como lo podemos probar? en este caso es facil verlo
v1 = np.array([ 1, 1])
v2 = np.array([-1,-1])
print(v1)
print(v2)
```

```python
# si efectivamente,  v2 es -v1
print(v1 == -1*v2)
```

```python
# Aqui estamos trabajando solo con 2 vectores y pudimos verlo, pero en espacios mas grandes no siempre es facil
# Para formalizar, un conjunto de vectores es linealmente independiente si ninguno de ellos es una combinacion
# lineal de los otros.
```
