```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Hasta ahora vimos que podemos representar un sistema de ecuaciones lineales de forma matricial, utilizando
# matrices y vectores, esto nos permite resolver varias ecuaciones al mismo tiempo de una manera metodica
# Examinemos ahora distintos sistemas de ecuaciones lineales y aprendamos a reconocer cuando tendran solucion
# Lo haremos en R2 para poder graficarlo facilmente, pero las conclusiones que obtengamos tambien aplicaran a 
# mas dimensiones
```

```python
# Comencemos con 3 ecuaciones y 2 incognitas
# y =  3*x+5
# y = -1*x+3
# y =  2*x+1

x = np.arange(-6,6)
y_1 =  3*x+5
y_2 = -1*x+3
y_3 =  2*x+1

plt.figure()
plt.plot(x,y_1)
plt.plot(x,y_2)
plt.plot(x,y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

#dibujamos los ejes
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')
plt.show()
```

```python
# Graficamente vemos que cuando tenemos mas ecuaciones que incognitas el sistema no tiene solucion. Como vemos
# que no tiene solucion? para que un punto sea solucion tiene que cumplir las 3 ecuaciones, o sea necesitamos
# un punto en el grafico que pertenezca al mismo tiempo a las 3 rectas, y como vemos no hay.

# Si escribimos como en la clase anterior las ecuaciones en forma matricial
# -3*x + y = 5
#    x + y = 3
# -2*x + y = 1

A = np.array([[-3,1],[1,1],[-2,1]])
print(A)
```

```python
# Ahora usemos 2 ecuaciones y 2 incognitas
# y = -1*x+3
# y =  2*x+1

x = np.arange(-6,6)
y_2 = -1*x+3
y_3 =  2*x+1

plt.figure()
plt.plot(x,y_2)
plt.plot(x,y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

#dibujamos los ejes
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')
plt.show()
```

```python
# Tiene solucion unica, y esta en linea con nuestra intuicion cuando 2 lineas se cruzan lo hacen en un unico punto
```

```python
# Ahora usemos 1 ecuaciones y 2 incognitas
# y =  2*x+1

x = np.arange(-6,6)
y_3 =  2*x+1

plt.figure()
plt.plot(x,y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

#dibujamos los ejes
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')
plt.show()
```

```python
# Lo que acabamos de ver son ejemplos de sistemas sobredeterminados, cuando tenemos mas ecuaciones que incognitas
# y por lo tanto no tienen solucion. Tambien vimos un sistema no determinado, es cuando solo tenemos 1 ecuacion 
# y 2 o mas incognitas, en este caso existe un numero infinito de soluciones (todos los puntos de la recta)

# Si pensamos todo esto en terminos de matrices, un sistema es sobredeterminado cuando tiene mas filas filas
# que columnas, o sea no es cuadrado.

# Y es subdeterminado cuando tiene mas columnas que filas, y por lo tanto tampoco es cuadrado.
```
