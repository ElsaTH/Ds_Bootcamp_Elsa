```python
def plotVectors(vecs, cols, alpha=1):
    """
    Plot set of vectors.

    Parameters
    ----------
    vecs : array-like
        Coordinates of the vectors to plot. Each vectors is in an array. For
        instance: [[1, 3], [2, 2]] can be used to plot 2 vectors.
    cols : array-like
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.
    alpha : float
        Opacity of vectors

    Returns:

    fig : instance of matplotlib.figure.Figure
        The figure of the vectors
    """
    plt.figure()
    plt.axvline(x=0, color='#A9A9A9', zorder=0)
    plt.axhline(y=0, color='#A9A9A9', zorder=0)

    for i in range(len(vecs)):
        x = np.concatenate([[0,0],vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=cols[i],
                  alpha=alpha)
```

```python
plotVectors([v1, v2], ['orange', 'blue'])
plt.xlim(-1, 8)
plt.ylim(-1, 8)
```

```python
# Y escribamos la combinacion lineal 2 veces v1 mas 1 vez v2
v1v2 = 2*v1+v2
plotVectors([v1, v2, v1v2], ['orange', 'blue', 'red'])
plt.xlim(-1, 20)
plt.ylim(-1, 20)
```

```python
# Lo que estamos haciendo es escribir a nuestro nuevo vector v1v2 en funcion de los vectores que ya teniamos
# v1 y v2, es como cuando tenemos que darle a alguien la direccion para ir de un lugar a otro, tenemos 
# muchas formas de hacerlo, pero si todos acordamos que lo haremos en funcion de cuadras y giros entonces luego
# podemos indicarle el camino como avanza 3 cuadras, gira a la derecha y camina 1 cuadra mas. Pero podriamos
# elegir otro sistema de referencia en lugar de cuadras y y giros, podriamos dar indicaciones en pasos y giros
# seria como elegir otros v1 y v2, podriamos obtener el mismo resultado? bueno no siempre, los vectores deben tener
# una propiedad para poder asegurar eso.
```

```python
# Definamos nuevamente a v1 y v2
v1 = [2, 5]
v2 = [3, 2]

# Grafiquemos una muestra del espacio generado  por los vectores v1 y v2
for a in range(-10, 10):
    for b in range(-10, 10):
        plt.scatter(v1[0] * a + v2[0] * b, v1[1] * a + v2[1] * b, marker='.', color='orange')

# Defining x and y sizes
plt.xlim(-100, 100)
plt.ylim(-100, 100)

# Draw axes
plt.axvline(x=0, color='#A9A9A9')
plt.axhline(y=0, color='#A9A9A9')

plt.show()
```
