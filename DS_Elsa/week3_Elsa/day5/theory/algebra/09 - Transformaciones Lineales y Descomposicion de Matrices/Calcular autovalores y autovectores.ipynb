```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Numpy nos da una funcion para poder calcular los autovalores y autovectores, la funcion es np.linalg.eig()
```

```python
# Usando el ejemplo de la clase anterior
X = np.array([[3,2], [4,1]])
print(X)
```

```python
# Calculemos para esa matriz los autovalores y autovectores
print(np.linalg.eig(X))
```

```python
# Nos devuelve un vector con los autovalores y una matriz con los autovectores
autovalores, autovectores = np.linalg.eig(X)
print(autovalores)
```

```python
# Esta matriz tiene 2 autovalores, el 5 y el -1
# El autovalor 5 tiene asociado el autovector 0.7 , 0.7
print(autovectores[:,0])
```

```python
# Mientras que el autovalor -1 tiene asociado el autovector -0.44, 0.89
print(autovectores[:,1])
```

```python
# Cuando calculamos los autovectores podemos obtener distintos vectores, porque no necesariamente tienen la misma
# escala, pero si deben tener la misma direccion.
```

```python
def plotVectors(vecs, cols, alpha=1):
    """
    Plot set of vectors.

    https://www.kaggle.com/robaks/learn-linear-algebra-for-data-science

    Parameters
    ----------
    vecs : array-like
        Coordinates of the vectors to plot. Each vectors is in an array. For
        instance: [[1, 3], [2, 2]] can be used to plot 2 vectors.
    cols : array-like
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.
    alpha : float
        Opacity of vectors

    Returns:

    fig : instance of matplotlib.figure.Figure
        The figure of the vectors
    """
    
    plt.axvline(x=0, color='#A9A9A9', zorder=0)
    plt.axhline(y=0, color='#A9A9A9', zorder=0)

    for i in range(len(vecs)):
        x = np.concatenate([[0,0],vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=cols[i],
                  alpha=alpha)
```

```python
v = np.array([[-1], [2]])
Xv = X.dot(v)
v_np = [-0.4472136, 0.89442719]

plotVectors([Xv.flatten(), v.flatten(), v_np], cols=['green', 'orange', 'blue'])
plt.ylim(-4, 2)
plt.xlim(-7, 3)
```
