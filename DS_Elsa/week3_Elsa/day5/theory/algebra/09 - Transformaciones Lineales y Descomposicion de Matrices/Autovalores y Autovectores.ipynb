```python
def plotVectors(vecs, cols, alpha=1):
    """
    Plot set of vectors.

    https://www.kaggle.com/robaks/learn-linear-algebra-for-data-science

    Parameters
    ----------
    vecs : array-like
        Coordinates of the vectors to plot. Each vectors is in an array. For
        instance: [[1, 3], [2, 2]] can be used to plot 2 vectors.
    cols : array-like
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.
    alpha : float
        Opacity of vectors

    Returns:

    fig : instance of matplotlib.figure.Figure
        The figure of the vectors
    """
    
    plt.axvline(x=0, color='#A9A9A9', zorder=0)
    plt.axhline(y=0, color='#A9A9A9', zorder=0)

    for i in range(len(vecs)):
        x = np.concatenate([[0,0],vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=cols[i],
                  alpha=alpha)
```

```python
#Definimos los colores que utilizaremos en nuestros graficos
naranja = '#FF9A13'
naranja_oscuro = 'orange'
azul = '#1190FF'
azul_oscuro = 'blue'
rojo = 'red'
verde = 'green'
```

```python
v = np.array([[1,1],[-2,2],[3,-6]])
origen = [0], [0] # punto de origen

plt.axvline(x=0, color='#A9A9A9', zorder=0)
plt.axhline(y=0, color='#A9A9A9', zorder=0)

plt.quiver(*origen, v[:,0], v[:,1], color=[rojo, azul_oscuro, verde], scale=20)
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()
```

```python
X = np.array([[-3,7], [5,5]])
print(X)
```

```python
v = np.array([[3], [2]])
print(v)
```

```python
plotVectors([v.flatten()], cols=[azul])
plt.xlim(-1,4)
plt.ylim(-1,4)
```

```python
u = X.dot(v)

plotVectors([v.flatten(), u.flatten()], cols=[azul, naranja])
plt.ylim(-1, 26)
plt.xlim(-1, 6)

print(u)
```
