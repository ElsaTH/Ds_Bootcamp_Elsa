```python
import os.path, sys

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder 
from sklearn.preprocessing import MinMaxScaler

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python


# encoder
le = LabelEncoder()
# Normalización:
def normalize_dataframe(df):
    scaler = MinMaxScaler()
    df_normalized = pd.DataFrame(
        data=scaler.fit_transform(df.values), 
        columns=df.columns, 
        index=df.index)
    return df_normalized

def clean_df(path=None):
    
    # Introducimos los datos del archivo en una variable común.
    path = os.path.abspath(input("Copie y pegue la dirección url o el copy path del archivo csv: "))
    # Creamos el dataframe
    df = pd.read_csv(path)
    # Eliminar todas las columnas con un solo valor, o con más de un 50% de valores faltantes, esto lo hacemos para trabajar más rápido.absDetermine si la              fila o columna se elimina de DataFrame, cuando tenemos al menos un NA o todos los NA.
    #Parámetros: 'all': si todos los valores son NA, elimine esa fila o columna.
    df.dropna(axis=0, how='all', inplace=True)
    # Eliminar duplicados
    df.drop_duplicates(keep=False, inplace=False)
    # Si hay pocos Nan y la columna puede servir podemos cambiar Nan por O
    df.fillna(0, inplace=True)
    print("NOMBRE KEYS PARA ELEGIR TARGET: ")
    print(df.keys())
    print("---------------------")
    # 2.Paso las columnas "Object" a numéricas
    X_categorical_no_numbers = df[df.select_dtypes('object').columns].apply(le.fit_transform)
    # 3. Cojo solo columnas numéricas
    X_others = df.select_dtypes(exclude=['object'])
    if len(X_others)>0:
        # 4. Normalizo las columnas numéricas
        X_others = normalize_dataframe(df=X_others)
    # 5. Concateno el resultado final
    df = pd.concat([X_others, X_categorical_no_numbers], axis=1)
    null_counts = df.isnull().sum()
    print("Number of null values in each column:\n{}".format(null_counts))
    print(df.keys())
    return df



def main():
    """
    Es la función que ha de ser ejecutada cada vez que queremos lanzar el programa.
    Al final, mostrará la precisión del modelo entrenado y retornará el modelo entreado.
    """
    # genero df llamando a la función
    df = clean_df(path=None)
    
    target_name = input("Elige columna como target")
    option_user = int(input("Elegir tipo de algoritmo: Regresión_lineal = 1, Regresión_logística = 2 o Knn = 3 "))
   
    if option_user == 1:  # modelo Regresión lineal
        params ={}
        params["n_jobs"] = int(input(" Cuántos procesos o subprocesos concurrentes quiere usar. n_jobs =  "))
        print("Tipo modelo elegido:  ", "Regresion lineal", "    Tipo de parámetro elegido  ", params) 
        print("---------------------")
        
    if option_user == 2: # modelo Regresion logística
        params ={}
        params["n_jobs"] = int(input("Cuántos procesos o subprocesos concurrentes quiere usar. n_jobs =  "))
        params["penalty"] = (input("Tipo parámetros que especifica la norma utilizada l1 o l2. Sino sabes cual por defecto penalty = l2 "))
        params["random_state"] = (input("Proporciona parámetro que controlar generador de números aleatorios. Por defecto random_state = None "))
        print("Tipo modelo elegido:  ","Regresion logistica", "    Tipo de parámetro elegido  ", params)
        print("---------------------")

    if option_user == 3: # modelo KNN
        params ={}
        params["n_jobs"] = int(input(" Cuántos procesos o subprocesos concurrentes quiere usar. n_jobs =  "))
        params["n_neighbors"] = int(input("Elige Número de vecinos kneighbors consultas. Prederteminado por skelear n_neigbors= 5  "))
        print("Tipo modelo elegido:  ","KNN", "    Tipo de parámetro elegido:  ", params)
        print("---------------------")

    # llamo a la función
    model = choose_model(option_user,params)
    model_trained, accuracy = train_model(model, df, target_name)
    
    print("Accuracy: ", accuracy*100, " %")
    return  model_trained


def choose_model(option_user, params):
    """
    Deberá crear el modelo necesario sin entrenar, retornándolo al final.
    """
    # Definimos el modelo
    if option_user == 1:  # modelo Regresión lineal
        model = LinearRegression()

    if option_user == 2: # modelo Regresion logística
        model = LogisticRegression()
      
    if option_user == 3: # modelo KNN
        model = KNeighborsClassifier()
  
    return model



def train_model(model, df, target_name):
    """
    Retornará el modelo entrenado con los datos y la precisión del modelo.
    """

    # Definimos X e y 
    X = df.drop(str(target_name),1).values.reshape(1, -1)
    y = df[str(target_name)].values

    # Realizamos el Split o la división de los datos de nuestro dataframe
    test_size = float((100 - int(input("Indique que porcentaje quiere aplicar al conjunto de entrenamiento en número entero. Ejemplo 80% --> 80")))/100)
    print("Porcentaje utilizado que dividie el entrenamiento es: ", (1-test_size)*100, " %")
    print("---------------------")

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_size)

   
    # Imprimimos por pantalla X_train e y_train para comprabar que todo está correcto.
    print("X_train:   ", X_train.shape)
    print("y_train:   ", y_train.shape)
    print("---------------------")

  
        
    # Entrenamos el modelo
    
    model_trained = model.fit(X_train, y_train)
    pred = model.predict(X_train)
    y_pred = model.predict(X_test)


    print("y_test = ", y_test.shape)
    print ("y_pred = ", y_pred.shape)
    print("---------------------")
    score_train = model.score(X_train, y_train) # Entrenamiento
    score_test = model.score(X_test, y_test)  # Test
    score_all = model.score(X, y)  # Todo
   
    

    # Visualización general del entrenamiento
    plt.plot(X_train, label="train", color="blue")
    plt.plot(X_test, label="test", color="orange")
    plt.title("Visualización general del entremamiento")
    plt.legend()
    print("---------------------")
    
    print("El conjunto de entrenamiento tiene un ", score_train *100, " %") 
    print("El conjunto de test tiene un ", score_test *100, " %")
    

    accuracy = model_trained.score(X_test, y_test)
    #accuracy = model_trained.accuracy_score(y_test, pred)--> No me permite el atributo accuracy_score con linearRegression
    #print("Accuracy:", metrics.accuracy_score(y_test, y_pred),normalize=False)
    return model_trained, accuracy
```
