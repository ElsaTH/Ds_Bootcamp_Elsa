A partir del archivo food_100, se pide:

1. Limpiar el archivo borrando las columnas Unnamed.
2. ¿Qué porcentaje de valores NaN hay en cada columna?
3. ¿Tendría algún sentido clasificar el nombre de los alimentos a partir del top5 de columnas numéricas con menos valores NaN?

```python
import pandas as pd
import numpy as np 
df = pd.read_csv("/Users/elsa/Desktop/Bootcamp The Bridge/The_Bridge/Ds_Bootcamp_Elsa/DS_Elsa/week9/day2/data/food_100.csv")

df
```

### Limpiar el archivo borrando las columnas Unnamed.




```python
# Compruebo cuantas columnas empiezan por Unnamed:
filter_col = [col for col in df if col.startswith('Unnamed')]
filter_col
```

```python
# Elimino la columna
df.drop('Unnamed: 0',1, inplace=True)
df
```

### ¿Qué porcentaje de valores NaN hay en cada columna?


```python
#Primero creamos la variable que los cuenta con isnull().sum()
missing_values_count = df.isnull().sum()
missing_values_count
```

```python
total_cells = np.product(df.shape)
total_cells
```

```python

#Contamos el total de valores missing, sumando todos los valores de la variable que contiene los missing para cada columna:
total_missing = missing_values_count.sum()
total_missing
```

```python
#El %:
total_missing/total_cells*100
```
